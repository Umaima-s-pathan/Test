```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##Pandas##

```python
data = {
'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
'Age': [24, 27, 22, 32, 29],
'Score': [85.5, 90.3, 88.0, 92.1, 78.8]
}
```

```python
# Create a DataFrame from the following dictionary
df=pd.DataFrame(data)
df
```

```python
# Using the DataFrame created in question 1, display the first three rows.
df.head(3)
```

```python
# Select and display the 'Name' and 'Score' columns from the DataFrame created in question 1.
cols=['Name','Score']
df[cols]
```

```python
# Filter the DataFrame to show only the rows where the 'Age' is greater than 25.
a=df[df['Age']>25]
print('names of those people with age greater than 25 \n-------------------------------------------\n')
a
```

```python
# Add a new column 'Passed' to the DataFrame, where the value is True if the 'Score' is greater than 80, otherwise False.
df['Passed']=df['Score']>80
df
```

```python
# Sort the DataFrame by the 'Score' column in descending order.
df.sort_values(by='Score',ascending=False)
```

```python
# Group the DataFrame by the 'Passed' column and calculate the mean 'Score' for each group.
a=df.groupby('Passed')['Score'].mean() # calculating mean of scores by grouping passed col
a
```

```python
# Add a new column 'Attendance' with values [90, np.nan, 85, 87, np.nan]. Write the code to fill missing values with the mean of the column.
d=(90,np.nan,85,87,np.nan)
df['Attendance']=pd.DataFrame(d)
df
```

```python
mean_atten=df['Attendance'].mean() # calculating mean value of atten col
df['Attendance'].fillna(value=mean_atten, inplace=True) # filling null values with mean value
df
```

##Data visualization

```python
#create a box plot using seaborn
data=[12, 15, 14, 10, 8, 11, 14, 18, 22, 24, 30, 28, 26, 30, 32, 35, 38]
sns.boxplot(data=data)
plt.title('Box Plot of Given Data \n--------------------------------------------\n')
plt.show()
```

```python
# create a histogram using matplotlib.
plt.hist(data, bins=10, edgecolor='black') # bins is for size of each bins of the histogram and
                                                  # edgecolor provide frames for each bin for better understanding

plt.title('histogram Plot of Given Data \n--------------------------------------------\n')
plt.show()
```

```python
# create a bar plot to show the 'Score' of each individual using seaborn.
df = pd.DataFrame({
    'Individual': range(1, len(data) + 1),  # Create labels for each individual
    'Score': data
})
sns.barplot(x='Individual', y='Score', data=df, palette='viridis') # pallete define the colors of the bars
plt.title('Bar Plot of Scores for Each Individual \n -----------------------------------------------------------\n')
plt.xlabel('Individual') # labelling the x axis
plt.ylabel('Score')  # labelling y axis
plt.show()
```

```python
#  Create a pie plot using the following data: ['Apples', 'Bananas', 'Cherries', 'Dates'] with corresponding values [10, 15, 7, 3].
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [10, 15, 7, 3] # corresponding data
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']  # Colors for each segment
explode = (0.1, 0, 0, 0)  # highlight the 1st slice (i.e., 'Apples')
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.2f%%', shadow=True) #autopct shows the percentage  value of each segment
                                                                                                # shadow creates shadow effect for 3D look
plt.title('Fruit Distribution')
plt.show()
```
