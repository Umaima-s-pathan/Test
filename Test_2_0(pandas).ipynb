```python
 #Q . Write a function `is_prime(num)` that checks if a given number is a prime number. Return `True` if the number is prime

def is_prime(n):
    i = 2

    if n in (1,3):
        return True
    elif n == 2:
        return False

    while i <= n//2:
        if n%i != 0:
            return True
        else:
            return False
        i+=1

for x in range(1, 5):
  print(x, '=', is_prime(x))
```
